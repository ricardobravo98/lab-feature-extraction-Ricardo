```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customer_df = pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv")
customer_df
```

```python
customer_df.columns
```

### Dropping Customer

```python
customer_df2 = customer_df.drop(columns = "Customer")
```

```python
numerical_df = customer_df2.select_dtypes("number")
categorical_df = customer_df2.select_dtypes(exclude = "number")
```

### Here I am plotting Bar plots for the categorical columns, I was thinking of dropping the dates but I consider it could be useful to see what dates occur more often than others

```python
for column in categorical_df.columns:
    if categorical_df[str(column)].nunique()<= 10:
        plt.figure(figsize=(12, 6))
        sns.countplot(x=column, data=categorical_df)
        plt.show()

    else:
        plt.figure(figsize=(12, 6))
        sns.countplot(y=column, data=categorical_df)
        plt.show()
```
